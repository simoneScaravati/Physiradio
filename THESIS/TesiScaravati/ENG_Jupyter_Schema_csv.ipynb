```python
import pandas as pd
import matplotlib.pyplot as plt

phys = pd.read_csv("ENG_Schema_articles_classification.csv")
phys
```

```python
print(phys.SIGHT.value_counts().sort_index())
phys.SIGHT.value_counts().sort_index().plot.bar()
plt.title("SIGHT")
plt.show()

print(phys.SIGHT.count())
```

```python
print(phys.TOUCH.value_counts().sort_index())
phys.TOUCH.value_counts().sort_index().plot.bar()
plt.title("TOUCH")
plt.show()

print(phys.TOUCH.count())
```

```python
print(phys.SMELL.value_counts().sort_index())
phys.SMELL.value_counts().sort_index().plot.bar()
plt.title("SMELL")
plt.show()

print(phys.SMELL.count())
```

```python
print(phys.HEARING.value_counts().sort_index())
phys.HEARING.value_counts().sort_index().plot.bar()
plt.title("HEARING")
plt.show()
print(phys.HEARING.count())

print(phys.TASTE.value_counts().sort_index())
phys.TASTE.value_counts().sort_index().plot.bar()
plt.title("TASTE")
plt.show()
print(phys.TASTE.count())
```

```python
# EXAMPLE  
# data = {'Name':['Tom', 'Jack', 'nick', 'juli'], 'marks':[99, 98, 95, 90]} 
# Creates pandas DataFrame. 
# df = pd.DataFrame(data, index =['rank1', 'rank2', 'rank3', 'rank4']) 

#0 e 1 
data = {'0':[ len(phys[phys.SIGHT == 0]), 
              len(phys[phys.TOUCH == 0]),
              len(phys[phys.SMELL == 0]),
              len(phys[phys.HEARING == 0]),
              len(phys[phys.TASTE == 0])] , 
        '1':[ len(phys[phys.SIGHT == 1]), 
              len(phys[phys.TOUCH == 1]),
              len(phys[phys.SMELL == 1]),
              len(phys[phys.HEARING == 1]),
              len(phys[phys.TASTE == 1])] } 
custom = pd.DataFrame(data, index=["SIGHT" , "TOUCH" , "SMELL" , "HEARING" , "TASTE"] )
custom
```

```python
custom.plot.bar()
#plt.show()
plt.xticks(rotation=0)
plt.savefig('ENG_barplot_full.png')
```
