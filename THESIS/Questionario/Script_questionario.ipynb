```python
#abbiamo bisogno dei vari pool di valori possibili --> liste
# ne faccio 8 : 7 per il mapping corretto e uno per mettere tutti i valori che non hanno senso

rilevati = [quest.DATI_PERCEPITO_1, quest.DATI_PERCEPITO_2, 
            quest.DATI_PERCEPITO_3, quest.DATI_PERCEPITO_4, 
            quest.DATI_PERCEPITO_5, quest.DATI_PERCEPITO_6, 
            quest.DATI_PERCEPITO_7]

valori_possibili = []
for valueOut in rilevati:
    for valueIn in valueOut:
        data2 = valueIn.split(":")
        for single2 in data2:
            s2 = single2.strip()
            valori_possibili.append(s2)

df_valori = pd.DataFrame(valori_possibili, columns=['valori_percepiti'])
df_valori.valori_percepiti.unique()
```

```python
#faccio dei set per i check
#PRIMA INIEZIONE DI ARBITRARIETÁ

mapping1_polluted = {'Sabbia', 'Fumo', 'Caligine', 'Caligine>85', 'Sabbia>85' , 'Fumo>85'}

mapping2_clear_humid   = {'Soleggiato Caldo Umido', 'Soleggiato Caldo e Umido', 'Sereno>85'}

mapping3_rainy_foggy  = {'Parzialmente Nuvoloso', 'Leggera Nebbia', 'Foschia'}

mapping4_clear  = {'Sereno', 'Soleggiato','Tramonto Soleggiato', 'Soleggiato Caldo',
                    'Sereno (notte)', 'Tramonto Sereno'}

mapping5_dangerous = {'Uragano','Temporale Estivo', 'Temporale','Grandine',
                    'Tornado', 'Acquazzone', 'Bufera', 'Nubifragio', 'Temporale forte', 
                    'Temporale estivo', 'Tromba d aria', 'Tempesta', 'Polvere' ,'Cenere',
                     'Cenere>85',  'Polvere>85', 'Tornado>85'}

mapping6_rainy_foggy_humid  = {'Pioggerella','Pioggia', 'Nuvoloso','Nebbia', 'Pioggia forte', 'Pioggerella>85',
                             'Foschia>85', 'Pioggia>85', 'Nebbia>85', 'Nuvoloso>85'}

mapping7_snow    = {'Neve', 'Neve forte', 'Nevischio', 'Rovesci nevosi'}

mapping_unknown  = {'Alba', 'Sera(?)', 'Tramonto(?)', 
                    'Buio cittadino', '-' , 'Fine pioggia(?)',
                    'Tramonto invernale sereno(?)', 'Fresco', 'Ventoso', 
                    'Vento', 'Prima della tempesta', 'Vento leggero',
                    'Caldo', 'Brezza', 'Vento forte', 'Brina' }
#unknown perché non esiste quella condizione meteo (o proprio non é una condizione meteo)
```

```python
#codice per vedere se ci sono dei valori non mappati
list_of_unmapped = []
all_values = df_valori.valori_percepiti.unique()

for i in all_values:
    if not(i in mapping1_polluted or i in mapping2_clear_humid or i in mapping3_rainy_foggy or 
            i in mapping4_clear or i in mapping5_dangerous or 
            i in mapping6_rainy_foggy_humid or i in mapping7_snow or i in mapping_unknown):
        list_of_unmapped.append(i)

list_of_unmapped
```

```python
#per il numero delle corrispondenze totali mi basterá fare
#la somma dei risultati ottenuti dai vari calcolaCorrispondenze e aggiungerli al df


#funzione per ogni campo della colonna restituisce una lista con
#1 se vi é una corrispondenza, 0 altrimenti

def column_to_corrispondenze(colonna, mappingSet):
    corrispondenze = []
    for x in colonna:
        data = x.split(":")
        result = 0
        for single in data:
            s = single.strip()
            if s in mappingSet:
                result = 1
        corrispondenze.append(result)
    return corrispondenze


risultati = pd.DataFrame(columns=['VERSIONE_MAPPING',
                                  'CORRISPONDENZA_DATO_1',
                                  'CORRISPONDENZA_DATO_2',
                                  'CORRISPONDENZA_DATO_3',
                                  'CORRISPONDENZA_DATO_4',
                                  'CORRISPONDENZA_DATO_5',
                                  'CORRISPONDENZA_DATO_6',
                                  'CORRISPONDENZA_DATO_7', 
                                  'TOTALE_CORRISPONDENZE_MAX_7'])

risultati.VERSIONE_MAPPING = quest.VERSIONE_MAPPING
risultati.CORRISPONDENZA_DATO_1 = column_to_corrispondenze(quest.DATI_PERCEPITO_1, mapping1_polluted)
risultati.CORRISPONDENZA_DATO_2 = column_to_corrispondenze(quest.DATI_PERCEPITO_2, mapping2_clear_humid)
risultati.CORRISPONDENZA_DATO_3 = column_to_corrispondenze(quest.DATI_PERCEPITO_3, mapping3_rainy_foggy)
risultati.CORRISPONDENZA_DATO_4 = column_to_corrispondenze(quest.DATI_PERCEPITO_4, mapping4_clear)
risultati.CORRISPONDENZA_DATO_5 = column_to_corrispondenze(quest.DATI_PERCEPITO_5, mapping5_dangerous)
risultati.CORRISPONDENZA_DATO_6 = column_to_corrispondenze(quest.DATI_PERCEPITO_6, mapping6_rainy_foggy_humid)
risultati.CORRISPONDENZA_DATO_7 = column_to_corrispondenze(quest.DATI_PERCEPITO_7, mapping7_snow)

risultati.TOTALE_CORRISPONDENZE_MAX_7 = risultati.sum(axis=1)
risultati
```
