```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

#dataframe questionari
quest = pd.read_csv("dati_questionario.csv")
quest
```

```python
tot_questionari = len(quest)
print("Totale questionari ", tot_questionari)
```

```python
quest.SESSO_M_F.value_counts().sort_index().plot.bar()
plt.title("Sesso del campione")
plt.xticks(rotation=0)
plt.show()
print(quest.SESSO_M_F.value_counts())
```

```python
quest.ETA.plot.hist()
plt.title("Eta")
plt.xticks(rotation=0)
plt.show()
#print(quest.ETA.value_counts(normalize=True))
print("MEAN" , quest.ETA.mean())
print("STD " , quest.ETA.std())
```

```python
quest.OCCUPAZIONE.value_counts().plot.bar()
plt.title("Occupazione")
plt.xticks(rotation=90)
plt.show()
quest.OCCUPAZIONE.value_counts(normalize=True)
# 71% studenti 
```

```python
quest.CITTA_RESIDENZA.value_counts().plot.bar()
plt.title("Citta")
plt.xticks()
plt.show()
quest.CITTA_RESIDENZA.value_counts(normalize=True)
```

```python
quest.TITOLO_STUDIO.value_counts().plot.bar()
plt.title("Titolo di studio")
plt.xticks(rotation=0)
plt.show()
quest.TITOLO_STUDIO.value_counts()
```

```python
quest.ASCOLTATORE_ABITUALE_SI_NO.value_counts().plot.bar()
plt.title("Ascoltatore abituale")
plt.xticks(rotation=0)
plt.show()
#quest.ASCOLTATORE_ABITUALE_SI_NO.value_counts()
quest.ASCOLTATORE_ABITUALE_SI_NO.value_counts(normalize=True)
```

```python
quest.EFFICACIA.value_counts().sort_index().plot.bar()
plt.title("Efficacia")
plt.xticks(rotation=0)
plt.show()
print(quest.EFFICACIA.value_counts(normalize=True))
print(quest.EFFICACIA.value_counts())
print("count" ,quest.EFFICACIA.count())
print("mean" , quest.EFFICACIA.mean())
print("median", quest.EFFICACIA.median())
print("std" , quest.EFFICACIA.std())
```

```python
quest.INCURIOSITO.value_counts().sort_index().plot.bar()
plt.title("INCURIOSITO")
plt.xticks(rotation=0)
plt.show()
print(quest.INCURIOSITO.value_counts(normalize=True))
print(quest.INCURIOSITO.value_counts())
print("mean" , quest.INCURIOSITO.mean())
print("median" , quest.INCURIOSITO.median())
print("std" , quest.INCURIOSITO.std())
```
