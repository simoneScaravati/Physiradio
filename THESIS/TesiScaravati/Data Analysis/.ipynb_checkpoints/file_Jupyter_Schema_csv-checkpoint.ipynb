```python
import pandas as pd
import matplotlib.pyplot as plt

phys = pd.read_csv("Schema_classificazione_articoli.csv")
phys
```

```python
print(phys.VISTA.value_counts().sort_index())
phys.VISTA.value_counts().sort_index().plot.bar()
plt.title("VISTA")
plt.show()

print(phys.VISTA.count())
```

```python
print(phys.TATTO.value_counts().sort_index())
phys.TATTO.value_counts().sort_index().plot.bar()
plt.title("TATTO")
plt.show()

print(phys.TATTO.count())
```

```python
print(phys.OLFATTO.value_counts().sort_index())
phys.OLFATTO.value_counts().sort_index().plot.bar()
plt.title("OLFATTO")
plt.show()

print(phys.OLFATTO.count())
```

```python
print(phys.UDITO.value_counts().sort_index())
phys.UDITO.value_counts().sort_index().plot.bar()
plt.title("UDITO")
plt.show()
print(phys.UDITO.count())

print(phys.GUSTO.value_counts().sort_index())
phys.GUSTO.value_counts().sort_index().plot.bar()
plt.title("GUSTO")
plt.show()
print(phys.GUSTO.count())
```

```python
# EXAMPLE  
# data = {'Name':['Tom', 'Jack', 'nick', 'juli'], 'marks':[99, 98, 95, 90]} 
# Creates pandas DataFrame. 
# df = pd.DataFrame(data, index =['rank1', 'rank2', 'rank3', 'rank4']) 

#0 e 1 
data = {'0':[ len(phys[phys.VISTA == 0]), 
              len(phys[phys.TATTO == 0]),
              len(phys[phys.OLFATTO == 0]),
              len(phys[phys.UDITO == 0]),
              len(phys[phys.GUSTO == 0])] , 
        '1':[ len(phys[phys.VISTA == 1]), 
              len(phys[phys.TATTO == 1]),
              len(phys[phys.OLFATTO == 1]),
              len(phys[phys.UDITO == 1]),
              len(phys[phys.GUSTO == 1])] } 
custom = pd.DataFrame(data, index=["VISTA" , "TATTO" , "OLFATTO" , "UDITO" , "GUSTO"] )
custom

#pd.crosstab(phys.VISTA , columns=["freq.ass"])
```

```python
custom.plot.bar()
#plt.show()
plt.xticks(rotation=0)
plt.savefig('barplot_full.png')
```
